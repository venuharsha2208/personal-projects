```python
from sklearn.preprocessing import StandardScaler
from hpelm import ELM
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.model_selection import train_test_split

# Assuming you have a DataFrame 'stock_data' with columns 'Output_1', 'Output_2', and 'Output_3' as target variables
# Features and target variable
X = stock_data.drop(columns=['Output_1', 'Output_2', 'Output_3','Output_4'], axis=1)  # Exclude the target
y = stock_data[['Output_1', 'Output_2', 'Output_3','Output_4']]

# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Step 4: Create and train the ELM model for each output
model_output_1 = ELM(X_train.shape[1], 1)  # 1 output for regression
model_output_1.add_neurons(1000, 'sigm')
model_output_1.train(X_train, y_train['Output_1'].values)

model_output_2 = ELM(X_train.shape[1], 1)
model_output_2.add_neurons(1000, 'sigm')
model_output_2.train(X_train, y_train['Output_2'].values)

model_output_3 = ELM(X_train.shape[1], 1)
model_output_3.add_neurons(1000, 'sigm')
model_output_3.train(X_train, y_train['Output_3'].values)

model_output_4 = ELM(X_train.shape[1], 1)
model_output_4.add_neurons(1000, 'sigm')
model_output_4.train(X_train, y_train['Output_4'].values)

# Step 5: Make predictions for each output
predictions_output_1 = model_output_1.predict(X_test)
predictions_output_2 = model_output_2.predict(X_test)
predictions_output_3 = model_output_3.predict(X_test)
predictions_output_4 = model_output_4.predict(X_test)

# Step 6: Evaluate the model for each output
mse_1 = mean_squared_error(y_test['Output_1'], predictions_output_1)
mse_2 = mean_squared_error(y_test['Output_2'], predictions_output_2)
mse_3 = mean_squared_error(y_test['Output_3'], predictions_output_3)
mse_4 = mean_squared_error(y_test['Output_4'], predictions_output_4)

print(f'Mean Squared Error for Output_1: {mse_1}')
print(f'Mean Squared Error for Output_2: {mse_2}')
print(f'Mean Squared Error for Output_3: {mse_3}')
print(f'Mean Squared Error for Output_4: {mse_4}')

# Step 7: Make sample predictions
random_indices = np.random.choice(len(X_test), size=5, replace=False)

sample_actual_values = y_test.iloc[random_indices]
sample_predictions_output_1 = predictions_output_1[random_indices]
sample_predictions_output_2 = predictions_output_2[random_indices]
sample_predictions_output_3 = predictions_output_3[random_indices]
sample_predictions_output_4 = predictions_output_4[random_indices]
print("-" * 150)
# Print the actual and predicted values
print(f"{'Sl':<10}{'1 Day actual':<15}{'1 day pred':<15}{'3 days actual':<15}{'3 days pred':<15}{'5 days actual':<15}{'5 days pred':<15}{'7 days actual':<15}{'7 days pred':<15}")
print("-" * 150)
for i in range(len(random_indices)):
    print(f"{i + 1:<10}{sample_actual_values.iloc[i, 0]:<15.2f}{sample_predictions_output_1[i][0]:<15.2f}{sample_actual_values.iloc[i, 1]:<15.2f}{sample_predictions_output_2[i][0]:<15.2f}{sample_actual_values.iloc[i, 2]:<15.2f}{sample_predictions_output_3[i][0]:<15.2f}{sample_actual_values.iloc[i, 3]:<15.2f}{sample_predictions_output_4[i][0]:<15.2f}")
```

# SLFN

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Assuming stock_data is already loaded
X = stock_data.drop(columns=['Output_1', 'Output_2', 'Output_3','Output_4'])  # Exclude the target variables
y = stock_data[['Output_1', 'Output_2', 'Output_3','Output_4']].values  # Multi-output target

# Step 1: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Convert data to TensorFlow tensors
X_train = tf.convert_to_tensor(X_train, dtype=tf.float32)
X_test = tf.convert_to_tensor(X_test, dtype=tf.float32)
y_train = tf.convert_to_tensor(y_train, dtype=tf.float32)
y_test = tf.convert_to_tensor(y_test, dtype=tf.float32)

# Step 3: Define the SLFN model
def build_slfn(input_size, hidden_neurons, activation):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(hidden_neurons, activation=activation, input_shape=(input_size,)),
        tf.keras.layers.Dense(y_train.shape[1])  # Output layer with multiple neurons for multi-output
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='mse',
                  metrics=['mse'])
    return model

# Step 4: Hyperparameter tuning
param_grid = {
    'hidden_neurons': [100, 200, 300],  # Number of neurons
    'activation': ['sigmoid', 'tanh']  # Activation functions
}

best_model = None
best_mse = float('inf')
best_r2 = float('-inf')

for hidden_neurons in param_grid['hidden_neurons']:
    for activation in param_grid['activation']:
        print(f"Training with {hidden_neurons} neurons and {activation} activation...")
        
        # Build and train the model
        model = build_slfn(X_train.shape[1], hidden_neurons, activation)
        history = model.fit(X_train, y_train, epochs=500, batch_size=32, verbose=0)
        
        # Evaluate the model
        test_predictions = model.predict(X_test, verbose=0)
        
        # Calculate Mean Squared Error and R² for each output
        mse = mean_squared_error(y_test.numpy(), test_predictions)
        r2 = r2_score(y_test.numpy(), test_predictions)
        print(f'Mean Squared Error: {mse}, R²: {r2}')
        
        # Keep track of the best model
        if mse < best_mse:
            best_mse = mse
            best_r2 = r2
            best_model = model
            best_history = history

# Step 5: Use the best model for predictions
final_predictions = best_model.predict(X_test, verbose=0)

# Print final model details
print(f'Best Model MSE: {best_mse}, Best Model R²: {best_r2}')

# Step 6: Make sample predictions
random_indices = np.random.choice(len(X_test), size=5, replace=False)
sample_actual_values = y_test.numpy()[random_indices]
sample_predictions = final_predictions[random_indices]
print("-" * 150)
# Print the actual and predicted values for each output
print(f"{'Sample':<10}{'1 day Actual':<15}{'1 day Predicted':<15}{'3 day Actual':<15}{'3 day Predicted':<15}{'5 day Actual':<15}{'5 day Predicted':<15}{'7 day Actual':<15}{'7 day Predicted':<15}")
print("-" * 150)
for i in range(len(random_indices)):
    print(f"{i + 1:<10}{sample_actual_values[i][0]:<15.2f}{sample_predictions[i][0]:<15.2f}{sample_actual_values[i][1]:<15.2f}{sample_predictions[i][1]:<15.2f}{sample_actual_values[i][2]:<15.2f}{sample_predictions[i][2]:<15.2f}{sample_actual_values[i][3]:<15.2f}{sample_predictions[i][3]:<15.2f}")

# Step 7: Plot training loss curve for the best model
plt.figure(figsize=(10, 6))
plt.plot(best_history.history['loss'], label='Training Loss (MSE)', color='blue', linewidth=2)
plt.title('Training Loss Curve for Best Model', fontsize=16)
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('Mean Squared Error (Loss)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

# Step 8: Plot Actual vs Predicted for Test Set
plt.figure(figsize=(12, 8))

# Plot actual vs predicted for each output
for i in range(4):  # For each output
    plt.subplot(2, 2, i + 1)
    plt.scatter(y_test.numpy()[:, i], final_predictions[:, i], alpha=0.6)
    plt.plot([y_test.numpy()[:, i].min(), y_test.numpy()[:, i].max()],
             [y_test.numpy()[:, i].min(), y_test.numpy()[:, i].max()],
             'r--', lw=2)
    plt.xlabel(f'Actual Output {i + 1}')
    plt.ylabel(f'Predicted Output {i + 1}')
    plt.title(f'Actual vs Predicted - Output {i + 1}')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import r2_score

# Assuming your DataFrame is named 'filtered_df'

# Define days and corresponding output columns
days = [1, 3, 5, 7]
output_columns = ['Output_1', 'Output_2', 'Output_3', 'Output_4']
model_names = ['ELM', 'slfn', 'ABC_ELM']

# Create a dictionary to store accuracy results
accuracy_results = {}

# Calculate accuracy for each model and output
for model in model_names:
    accuracy_results[model] = {}
    for i in range(len(days)):
        actual_column = output_columns[i]
        predicted_column = f'{model}_out{i + 1}'

        # Calculate R-squared
        r2 = r2_score(filtered_df[actual_column], filtered_df[predicted_column])

        accuracy_results[model][days[i]] = r2

# Print the accuracy results
for model, results in accuracy_results.items():
    print(f"Accuracy for {model}:")
    for day, accuracy in results.items():
        print(f"  Day {day}: {accuracy:.4f}")
```
